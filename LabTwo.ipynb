#Lab 2: Basics Operations on Image

# **Task 1: Read the image Mario and display it like this**

```python
from PIL import Image
from IPython.display import display
import sys

try:
    img = Image.open("/content/Mario.png")
    display(img)

except IOError:
    print("Unable to load image")
    sys.exit(1)
```

# **Task 2: Convert the same image into 2nd display image.**

```python
flipped_img = img.transpose(Image.FLIP_TOP_BOTTOM)
display(flipped_img)
```

# **Task 3: Convert the Mario image into Gray Scale.**

```python
gray_image = img.convert('L')
display(gray_image)
```
